```python
import matplotlib.pyplot as plt

class SecantMethod:
    def __init__(self, func, x0, x1, tol=1e-6, max_iter=100):
        self.func = func
        self.x0 = x0
        self.x1 = x1
        self.tol = tol
        self.max_iter = max_iter

    def solve(self):
        iteration = 0
        approximations = []  # To store the approximations at each iteration

        while iteration < self.max_iter:
            f_x0 = self.func(self.x0)
            f_x1 = self.func(self.x1)

            if abs(f_x1) < self.tol:
                return self.x1, approximations

            denominator = f_x1 - f_x0
            if abs(denominator) < 1e-12:
                raise ValueError("Secant method: Division by zero")

            x_next = self.x1 - (f_x1 * (self.x1 - self.x0) / denominator)

            if abs(x_next - self.x1) < self.tol:
                return x_next, approximations

            self.x0, self.x1 = self.x1, x_next
            approximations.append(x_next)
            iteration += 1

        raise ValueError("Secant method: Maximum number of iterations reached")

# Define the function you want to find the root of
def f(x):
    return (x*x/1) +((x) *4)

if __name__ == "__main__":
    # Initial guesses
    x0 = 1.0
    x1 = 2.0

    secant_solver = SecantMethod(func=f, x0=x0, x1=x1)
    root, approximations = secant_solver.solve()

    print(f"Approximate root: {root}")

    # Plot the approximations
    plt.plot(approximations, marker='o', linestyle='-')
    plt.xlabel('perulangan')
    plt.ylabel('nilai perkiraan')
    plt.title('konvergensi dari Secant Method')
    plt.grid(True)
    plt.show()
```
